```python
import pandas as pd
import numpy as np
data = pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv")
```

### Features

```python
data = data[['Make',
'Model',
'Year',
'Engine HP',
'Engine Cylinders',
'Transmission Type',
'Vehicle Style',
'highway MPG',
'city mpg',]]
```

### Data preparation

```python
data.columns = data.columns.str.replace(' ', '_').str.lower()
```

```python
data.fillna(0)
```

```python
data.rename(columns= {'MSRP':'price'})
```
