```python
import pandas as pd
import numpy as np
data = pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv")
data
```

### Features

```python
data = data[['Make',
'Model',
'Year',
'Engine HP',
'Engine Cylinders',
'Transmission Type',
'Vehicle Style',
'highway MPG',
'city mpg',
'MSRP']]
```

### Data preparation

```python
data.columns = data.columns.str.replace(' ', '_').str.lower()
```

```python
data.fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.rename(columns= {'msrp':'price'})
```

```python
from sklearn.model_selection import train_test_split

full_train, data_test = train_test_split(data, test_size=0.2, random_state=42)
data_train, data_val = train_test_split(full_train, test_size=0.20, random_state=42)


data_train.reset_index(drop=True, inplace=True)
data_test.reset_index(drop=True, inplace=True)
data_val.reset_index(drop=True, inplace=True)


y_train = data_train['msrp'].values
y_test = data_test['msrp'].values
y_val = data_val['msrp'].values


del data_train['msrp']
del data_test['msrp']
del data_val['msrp']
```

```python
y_train_num = y_train.copy()
y_val_num = y_val.copy()
y_test_num = y_test.copy()

y_train = (y_train > y_train.mean()).astype(int)
y_val = (y_val > y_val.mean()).astype(int)
y_tes = (y_test > y_test.mean()).astype(int)
```

```python
data['msrp'].mean
```

```python
above_average = data['msrp'].apply(lambda x: 1 if data['msrp'].mean() < x else 0)
```

```python
from sklearn.metrics import mutual_info_score

for bam in kategorik:
    data[bam]
    print(bam)
    monster_score = mutual_info_score(above_average, data[bam])
    print(round(monster_score,3))
```

```python
from sklearn.feature_extraction import DictVectorizer

train_dicts = data_train[kategorik + numerik].to_dict(orient='records')

dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)

val_dicts = data_val[kategorik + numerik].to_dict(orient='records')
X_val = dv.transform(val_dicts)

test_dicts = data_test[kategorik + numerik].to_dict(orient='records')
X_test = dv.transform(test_dicts)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_val, model.predict(X_val))
round(acc, 2)
```
